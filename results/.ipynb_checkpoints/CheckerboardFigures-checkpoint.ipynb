```python
import matplotlib.pyplot as plt
import numpy as np
import sys
sys.path.append('..')
from util.experiment_util import *
```

```python
# Plotting configurations
import matplotlib
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['text.latex.unicode'] = True
SMALL_SIZE = 16
MEDIUM_SIZE = 22
BIGGER_SIZE = 24
plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels 
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=14)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

# Checkerboard 2

### Ground Truth Plot

```python
checker2_data = np.load('/model-change-paper/results/checker2-1/X_labels.npz')
X2, labels2 = checker2_data['X'], checker2_data['labels']
```

```python
gt_colors = labels2.copy().astype('str')
gt_colors[gt_colors == '1'] = 'r'
gt_colors[gt_colors == '-1'] = 'b'
plt.figure(figsize=(4,4))
plt.scatter(X2[:,0], X2[:,1], c=gt_colors)
plt.yticks([0,0.5,1.0])
plt.tight_layout()
plt.savefig('/model-change-paper/results/checker2.pdf')
plt.show()
```

### Sequential Test Plots

```python
acq_models = ['mc--gr', 'mc--log', 'mc--probitnorm', 'rand--gr', 'uncertainty--gr', 'uncertainty--log', \
              'uncertainty--probitnorm', 'vopt--gr', 'sopt--gr']
c2_root = './checker2-1/'
tau, gamma = 0.1, 0.1
cand, select_method = 'full', 'top'
M, B, num_al_iters = 50, 1, 200
run_nums = [1]
gt_colors = labels2.copy().astype('str')
gt_colors[gt_colors == '1'] = 'r'
gt_colors[gt_colors == '-1'] = 'b'

for am in acq_models:
    acq, modelname = am.split('--')
    print(am)
    runs = get_data_from_runs(acq, modelname, M, tau, gamma, cand, select_method, \
                              B, num_al_iters, runs=run_nums, root_filename=c2_root)
    init_labeled = runs[run_nums[0]]['init_labeled']
    choices = runs[run_nums[0]]['choices']
    print(len(init_labeled), len(choices))
    print(len(set(choices)))
    plt.figure(figsize=(4,4))
    plt.scatter(X2[:,0], X2[:,1], c=gt_colors, alpha=0.8)
    plt.scatter(X2[init_labeled,0], X2[init_labeled,1], marker='*', c='gold', s=90, edgecolors='k', linewidths=0.4)
    plt.scatter(X2[choices,0], X2[choices,1], marker='*', c='gold', s=90, edgecolors='k', linewidths=0.4)
    plt.tight_layout()
    plt.xticks([], [])
    plt.yticks([], [])
    plt.savefig('%s%s-%s.pdf'% (c2_root, acq, modelname))
    plt.show()
    
```

### Batch Test Plots

```python
acq_models = ['mc--gr', 'mc--log', 'mc--probitnorm', 'rand--gr', 'uncertainty--gr', 'uncertainty--log', \
              'uncertainty--probitnorm', 'vopt--gr', 'sopt--gr']
c2_root = './checker2-5/'
tau, gamma = 0.1, 0.1
cand, select_method = 'rand', 'top'
M, B, num_al_iters = 50, 5, 100
run_nums = [3]
gt_colors = labels2.copy().astype('str')
gt_colors[gt_colors == '1'] = 'r'
gt_colors[gt_colors == '-1'] = 'b'
total_points = 200

for am in acq_models:
    acq, modelname = am.split('--')
    print(am)
    runs = get_data_from_runs(acq, modelname, M, tau, gamma, cand, select_method, \
                              B, num_al_iters, runs=run_nums, root_filename=c2_root)
    init_labeled = runs[run_nums[0]]['init_labeled']
    choices = runs[run_nums[0]]['choices']
    print(len(init_labeled), len(choices))
    print(len(set(choices)))
    plt.figure(figsize=(4,4))
    plt.scatter(X2[:,0], X2[:,1], c=gt_colors, alpha=0.8)
    plt.scatter(X2[init_labeled,0], X2[init_labeled,1], marker='*', c='gold', s=90, edgecolors='k', linewidths=0.4)
    plt.scatter(X2[choices[:total_points],0], X2[choices[:total_points],1], marker='*', c='gold', s=90, edgecolors='k', linewidths=0.4)
    plt.tight_layout()
    plt.xticks([], [])
    plt.yticks([], [])
    plt.savefig('%s%s-%s.pdf'% (c2_root, acq, modelname))
    plt.show()
    
```

# Checkerboard 3 Plots

### Ground Truth Plot

```python
checker3_data = np.load('/model-change-paper/results/checker3/X_labels.npz')
X, labels = checker3_data['X'], checker3_data['labels']
```

```python
gt_colors3 = labels.copy().astype('str')
gt_colors3[gt_colors3 == '0'] = 'r'
gt_colors3[gt_colors3 == '1'] = 'b'
gt_colors3[gt_colors3 == '2'] = 'g'


plt.figure(figsize=(4,4))
plt.scatter(X[:,0], X[:,1], c=gt_colors3)
plt.yticks([0,0.33,0.67,1.0])
plt.xticks([0,0.33,0.67,1.0])
plt.tight_layout()
plt.savefig('/model-change-paper/results/checker3.pdf')
plt.show()
```

### Batch Test Plots

```python
acq_models = ['mc--gr', 'mcgreedy--ce', 'uncertainty--gr','uncertainty--ce', 'rand--gr', 'vopt--gr']
c3_root = './checker3/'
tau, gamma = 0.1, 0.1
tau_ce, gamma_ce = 0.05, 0.01
cand, select_method = 'rand', 'top'
M, B, num_al_iters = 50, 5, 100
run_nums = [4]
gt_colors3 = labels.copy().astype('str')
gt_colors3[gt_colors3 == '0'] = 'r'
gt_colors3[gt_colors3 == '1'] = 'b'
gt_colors3[gt_colors3 == '2'] = 'g'

for am in acq_models:
    acq, modelname = am.split('--')
    print(am)
    if modelname == 'ce':
        runs = get_data_from_runs(acq, modelname, M, tau_ce, gamma_ce, cand, select_method, \
                              B, num_al_iters, runs=run_nums, root_filename=c3_root)
    else:
        runs = get_data_from_runs(acq, modelname, M, tau, gamma, cand, select_method, \
                              B, num_al_iters, runs=run_nums, root_filename=c3_root)
    init_labeled = runs[run_nums[0]]['init_labeled']
    choices = runs[run_nums[0]]['choices']
    print(len(choices))
    print(len(set(choices)))
    plt.figure(figsize=(4,4))
    plt.scatter(X[:,0], X[:,1], c=gt_colors3, alpha=0.8)
    plt.scatter(X[init_labeled,0], X[init_labeled,1], marker='*', c='gold', s=90, edgecolors='k', linewidths=0.4)
    plt.scatter(X[choices[:250],0], X[choices[:250],1], marker='*', c='gold', s=90, edgecolors='k', linewidths=0.4)
    plt.tight_layout()
    plt.xticks([], [])
    plt.yticks([], [])
    #plt.savefig('./checker3/%s-%s.pdf'% (acq, modelname))
    plt.show()
    
    
```
